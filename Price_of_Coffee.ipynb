Import the Librabries



```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

Load the dataset

```python
dataset = pd.read_csv('coffee_prices_10x2.csv')
```

Summarize the dataset

```python
print(dataset.shape)
print(dataset.head(5))
```

Visualization of the dataset

```python
plt.xlabel('Size_oz')
plt.ylabel('Price_usd')
plt.scatter(dataset.Size_oz, dataset.Price_usd, color ='blue', marker = 'o')
```

Segregate The Dataset

```python
Size_oz = dataset.drop('Price_usd', axis = 'columns')
Size_oz
```

```python
Price_usd = dataset.Price_usd
Price_usd
```

Training the dataset using Linear Regression

```python
model = linear_model.LinearRegression()
model.fit(Size_oz, Price_usd)
```

Predicted Price for Size_oz of 25

```python
Size_oz = [[18]]
PredictedmodelResult = model.predict(Size_oz)
print(PredictedmodelResult)
```

Check if the model is right

Theory calculation

y = m*x + b (m is coefficient and b is intercept)

Coefficient - m

```python
m = model.coef_
print(m)
```

Intercept - b

```python
b = model.intercept_
print(b)
```

y = m*x + b

x is independent variable_input_size_oz

```python
y = m*18 + b
print("The Price of {0} Size_oz is: {1}".format(Size_oz,y))
```
