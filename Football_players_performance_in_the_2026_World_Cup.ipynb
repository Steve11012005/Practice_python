Find Percentage of players with 7 games match

1.dataset['Class'] = 1 7 match played

2.dataset['Class'] = 0 Less than 7 match played

```python
matches_played = 0
matches_played = dataset['matches_played']
for i in range(len(matches_played)):
  if matches_played[i]==1:
    matches_played = matches_played+1
  matches_played
  per_m = (matches_played/len(matches_played))*100
  print('Percentage of total players with seven games match in the dataset: ', per_m)
```

Find Percentage of players with less than seven match played

```python
matches_played = 0
matches_played = dataset['matches_played']
for i in range(len(matches_played)):
  if matches_played[i] == 0:
    matches_played = matches_played+1
  matches_played
  per_m = (matches_played/len(matches_played))*100
  print('Percentage of total players with seven games match in the dataset: ', per_m)
```

```python
x = dataset.drop(['matches_played', 'player_id', 'player_name', 'position', 'nationality'], axis=1) # Drop the target variable and non-numeric columns

y = dataset['matches_played']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

linear =LinearRegression()

linear.fit(xtrain, ytrain)
```

```python
y_pred = linear.predict(xtest)

table= pd.DataFrame({"Actual":ytest,"Predicted":y_pred})

table
```
